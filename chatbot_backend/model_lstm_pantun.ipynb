```python

def clean_pantun(teks):
    # Ganti \n atau \\n dengan token <BR>
    teks = teks.replace('\\n', ' <BR> ').replace('\n', ' <BR> ')
    teks = teks.lower()
    teks = re.sub(r'[^a-z0-9<>\s]', '', teks)  # hapus tanda baca tapi jaga <BR>
    teks = re.sub(r'\s+', ' ', teks)  # rapikan spasi berlebih
    return teks.strip()


# Terapkan fungsi ke kolom
dfm['clean_teks'] = dfm['teks'].apply(clean_pantun)

# Cek hasil
print(dfm[['tipe', 'teks', 'clean_teks']].head())
```

```python
from collections import Counter

# Tokenisasi ke list of words
tokenized_pantun = [line.split() for line in dfm['clean_teks']]

# Buat vocab (semua kata unik + jumlah kemunculan)
word_counts = Counter([word for line in tokenized_pantun for word in line])
word2idx = {word: idx + 2 for idx, (word, _) in enumerate(word_counts.most_common())}
word2idx['<PAD>'] = 0
word2idx['<UNK>'] = 1
idx2word = {idx: word for word, idx in word2idx.items()}

# Simpan jumlah vocab
vocab_size = len(word2idx)
```

```python
# Buat n-gram sequences: dari tokenized_pantun
sequences = []
for line in tokenized_pantun:
    token_ids = [word2idx.get(word, word2idx['<UNK>']) for word in line]
    for i in range(1, len(token_ids)):
        ngram = token_ids[:i+1]
        sequences.append(ngram)

# Contoh hasil sequence
print("Contoh sequence:", sequences[0])
```

```python
import torch

# Tentukan panjang maksimum sequence
max_seq_len = max([len(seq) for seq in sequences])

# Padding manual (kiri)
def pad_sequence(seq, max_len):
    return [0] * (max_len - len(seq)) + seq

# Pisah jadi X dan y
X_data = []
y_data = []

for seq in sequences:
    padded = pad_sequence(seq[:-1], max_seq_len - 1)
    target = seq[-1]
    X_data.append(torch.tensor(padded, dtype=torch.long))
    y_data.append(torch.tensor(target, dtype=torch.long))

# Jadi dataset PyTorch siap diload
from torch.utils.data import Dataset, DataLoader

class PantunDataset(torch.utils.data.Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = PantunDataset(X_data, y_data)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)
```

```python
import torch.nn as nn

class PantunLSTM(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_dim * 2, vocab_size)  # *2 karena bidirectional

    def forward(self, x):
        x = self.embedding(x)
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])  # ambil output terakhir
        return out
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = PantunLSTM(vocab_size=vocab_size, embed_dim=100, hidden_dim=128).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)
```

```python
epochs = 10

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for X_batch, y_batch in dataloader:
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss:.4f}")
```
